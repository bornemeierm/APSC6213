```python
##inits
import sympy;
import numpy;
import matplotlib.pyplot as plot
import math;
from mpl_toolkits.mplot3d import Axes3D as ax
%matplotlib inline
```

```python
#I.C.'s
L = 1.0                          #length of bar                         
dx = 1000.                        #number of steps along the bar in 


n = int(1000);                    #number of n terms to compute in infinite series
Xrange = linspace(0,1,dx)        #number of steps from 0 to L (number of spatial gridpoints along the bar) 
   
trange= [0,.001, .01, .05, .1, 1.0, 10, 100]
D = 1                            #Thermal diffusivity of stainless .000003352   ; we use 1 to speed it up
                                 #steel at 25degrees C   
    
u = numpy.zeros((dx,8,4))#Solution array: dx # rows, totaltime # columns
                                 #(for t=0,1,2,3,4), and 4 elements 
                                 #(Temp(T), position along bar (X), time(t), 
                                 #number of terms in series(n)
nrange = linspace(1,n,n)                
Tsum = numpy.linspace(1,n,n)
tempsum=numpy.ones((1,1))

#totaltime = 1                  
#timesteps = 101;       
#trange = linspace(0,totaltime,timesteps) 
```

```python
print(trange)
```

```python
j = 0
for t in trange:    
    i=0
    for X in Xrange:
        u[i,j,1] = X 
        u[i,j,2] = t
        u[i,j,3] = n
        k=0
        for n in nrange:
            Tsum[k] = (2/(L*(((2*n-1)*pi)/2*L)))*sin(((2*n-1)*pi/(2*L))*X)*exp(-D*((2*n-1)*pi/(2*L))**2*t)
            k+=1 
        u[i,j,0]= numpy.ndarray.sum(Tsum)  
        i+=1   
    j+=1       
```

```python
#t=0 plot
positiont0 = u[:,0,1]
temperaturet0 = u[:,0,0]

#t=.001 plot
positionplot2 = u[:,1,1]
temperatureplot2 = u[:,1,0]

#t=.01 plot
positionplot3 = u[:,2,1]
temperatureplot3 = u[:,2,0]

#t=.05 plot
positionplot4 = u[:,3,1]
temperatureplot4 = u[:,3,0]

#t=.1 plot
positionplot5 = u[:,4,1]
temperatureplot5 = u[:,4,0]

#t=1 plot
positionplot6 = u[:,5,1]
temperatureplot6 = u[:,5,0]

#t=10 plot
positionplot7 = u[:,6,1]
temperatureplot7 = u[:,6,0]

#t=100 plot
positionplot8 = u[:,7,1]
temperatureplot8 = u[:,7,0]

#fig1 = plt.figure()

plt.figure(figsize=(20,16));

plt.plot(positiont0,temperaturet0,'g', label='t=0.0')
plt.plot(positionplot2, temperatureplot2,'b', label='t=0.001')
plt.plot(positionplot3, temperatureplot3,'r', label='t=0.01')
plt.plot(positionplot4, temperatureplot4,'k', label='t=0.1')
plt.plot(positionplot5, temperatureplot5,'y', label='t=0.5')
plt.plot(positionplot6, temperatureplot6,'m', label='t=1.0')
plt.plot(positionplot7, temperatureplot7,'c', label='t=10.0')
plt.plot(positionplot8, temperatureplot8,'g', label='t=100.0')
plt.legend(loc='center right', numpoints = 1)
 
plt.grid(True);
plt.title('HW4 Problem 1d, Temp vs. Position Plot for n=1000, dx=1000');
plt.xlabel(r'Position (X)', fontsize=18);
plt.ylabel(r'Temp (T)', fontsize=18);
plt.ylim(-.1,1.1)
plt.xlim(0,1)
```
