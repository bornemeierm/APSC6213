```python
##inits
import sympy;
import numpy;
import matplotlib.pyplot as plot
import math;
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#I.C.'s
L = 1.0                          #length of bar                         
dx = 1000.                       #number of steps along the bar in 
                                 #the x-direction
n=int(500);                       #number of n's in solution

totaltime = 1
Xrange = linspace(0,1,dx)        #range of values from 0 to 1 
                                 #across the bar that we want to sample

trange = linspace(0,totaltime,totaltime)
D =  .000003352                  #Thermal diffusivity of stainless
                                 #steel at 25degrees C

u = numpy.zeros((dx,totaltime,4))#Solution array: 101 rows, 5 columns 
                                 #(for t=0,1,2,3,4), and 4 elements 
                                 #(Temp(T), position along bar (X), time(t), 
                                 #number of terms in series(n)

nrange = linspace(1,n,n)                

Tsum = numpy.linspace(1,n,n)
```

```python
i = 0
j = 0

for t in trange:    
    i=0
    for X in Xrange:
        u[i,j,1] = X 
        u[i,j,2] = t
        u[i,j,3] = n
        k=0
        for n in nrange:
            Tsum[k] = (4/pi)*(((-1)**(n+1))/(2*n-1))*\
            cos(((2*n-1)*pi*(X))/2*L)\
            *exp((-D*t*((2*n-1)**2)*(pi**2))/(4*L**2))
            #print(j,i,k)
            k+=1 
        u[i,j,0] = numpy.ndarray.sum(Tsum)
        
        i+=1
    j+=1    
```
