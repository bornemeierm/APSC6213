```python
import sympy;
from scipy.integrate import quad;
import numpy;
import matplotlib.pyplot as plt;
%matplotlib inline
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
T_0 = 2.;
T_L = 1.;


L = 10.;

nx = 10;
ny = 10;

dx = L/(nx)
dy = L/(ny)

x = numpy.linspace(0,L,nx)
y = numpy.linspace(0,L,ny)
nf = 200

Tsum = numpy.zeros(nf);

T = numpy.zeros((nx+1,ny+1));

tempsum=numpy.ones((1,1));




xp = numpy.linspace(0,nx,nx+1)
yp = numpy.linspace(0,ny,ny+1)

xp.shape
```

```python
for x in range(0,nx+1,1):
    for y in range(0,ny+1,1):
        for n in range(1,nf,1):
            Tsum[n-1] = T_0 - (2./(n*numpy.pi))*(T_L - T_0)*((-1.)**n-1.)/(numpy.sinh(n*numpy.pi))*numpy.sinh(n*numpy.pi*x/L)*numpy.sin(n*numpy.pi*y/L)
        tempsum[0] = numpy.ndarray.sum(Tsum)
        T[x,y] = tempsum[0]
        
Tadjusted = T - 396
```

```python
mx, my = numpy.meshgrid(xp,yp)
```

```python
plt.figure(figsize=(8,5))
plt.contourf(my,mx,Tadjusted,500)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.colorbar();
```
